```python
import tensorflow as tf
tf.__version__
```

```python
# Loading the Fashion MNIST dataset in Keras
mnist = tf.keras.datasets.fashion_mnist
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
```

```python
training_images.shape
```

```python
# Data representation - Displaying the fourth digit
digit = training_images[4]
import matplotlib.pyplot as plt
plt.imshow(digit, cmap = plt.cm.binary)
plt.show()
```

```python
# Data Normalization
training_images = training_images / 255.0
test_images = test_images / 255.0
```

```python
# The network architecture
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                     tf.keras.layers.Dense(512, activation=tf.nn.relu),
                                     tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
# The compilation step
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# Train Network
model.fit(training_images, training_labels, epochs=5, batch_size=128)
```

```python
# check that the model performs well on the test set, too
model.evaluate(test_images, test_labels)
```
